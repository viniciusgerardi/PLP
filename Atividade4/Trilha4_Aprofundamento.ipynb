## Exploração de dados
Utilizando os datasets de usuários de banda larga e conclusão do ensino primário, retirados do https://www.gapminder.org/

Importação e análise dos dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Usuários de banda larga
usuarios = pd.read_csv('broadband_subscribers_per_100_people.csv')

print(f'Dimensões do dataset: {usuarios.shape}\n')
print(f'Países: {usuarios.country.unique()}')
print(f'Anos: {usuarios.columns[1:]}')
print(f'\nTipos de varíaveis:\n{usuarios.dtypes}\n')
print('\nAmostra dos dados:')

display(usuarios.head())
display(usuarios.describe())


print('\nChecagem de valores nulos:')
print(usuarios.isnull().sum())
```

```python
# Conclusão ensino primário
estudantes = pd.read_csv('primary_completion_rate_total_percent_of_relevant_age_group.csv')

print(f'Dimensões do dataset: {estudantes.shape}\n')
print(f'Países: {estudantes.country.unique()}')
print(f'Anos: {estudantes.columns[1:]}')
print(f'\nTipos de varíaveis:\n{estudantes.dtypes}\n')
print('\nAmostra dos dados:')

display(estudantes.head())
display(estudantes.describe())


print('\nChecagem de valores nulos:')
print(estudantes.isnull().sum())
```

Temos uma grande quantidade de valores nulos nos dois datasets, precisamos levar isso em consideração na seleção dos dados para análise.

## Preparação dos dados
Focando a análise na Costa Rica, vamos filtrar e combinar as informações

```python
pais='Costa Rica'
```

```python
# Usuários de banda larga
usuariosFiltrado=usuarios[usuarios.country==pais]
DS=pd.melt(usuariosFiltrado, id_vars=['country'])
DS=DS.rename(columns={'variable':'year','value':'usuariosBL'})
DS.head()
```

```python
# Estudantes
estudantesFiltrado=estudantes[estudantes.country==pais]
DS2=pd.melt(estudantesFiltrado, id_vars=['country'])
DS2=DS2.rename(columns={'variable':'year','value':'estudantes'})
DS2.head()
```

```python
# Combinando os dados
DS=pd.merge(DS, DS2,on=['country','year'])
DS.head()
```

```python
# Checando valores nulos
print(DS.isnull().sum())
display(DS)
```

Temos um intervalo sem dados nulos entre 2003 e 2018. Vamos retirar os demais anos para termos uma análise mais apurada

```python
DSfiltrado = DS[ DS.year > '2002' ].reset_index(drop=True)
display(DSfiltrado)
```

## Visualização e análise

```python
plt.figure(figsize=(15,10))
sns.lineplot(x=DSfiltrado.year, y=DSfiltrado.usuariosBL, label='Usuários de banda larga')
sns.lineplot(x=DSfiltrado.year, y=DSfiltrado.estudantes, label='Estudantes concluindo ensino primário')
plt.title('Costa Rica: Usuários de banda larga x Estudantes concluindo ensino primário')

plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x=DSfiltrado.year, y=np.log(DSfiltrado.usuariosBL), label='Usuários de banda larga')
sns.lineplot(x=DSfiltrado.year, y=np.log(DSfiltrado.estudantes), label='Estudantes concluindo ensino primário')
plt.title('Costa Rica: Usuários de banda larga x Estudantes concluindo ensino primário (log)')

plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Normalizando os dados
for c in DSfiltrado[['usuariosBL','estudantes']]:
  DSfiltrado[c] = DSfiltrado[c] / DSfiltrado[c].max()
DSfiltrado.head()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x=DSfiltrado.year, y=DSfiltrado.usuariosBL, label='Usuários de banda larga')
sns.lineplot(x=DSfiltrado.year, y=DSfiltrado.estudantes, label='Estudantes concluindo ensino primário')
plt.xlabel('Ano')
plt.ylabel('%')
plt.title('Costa Rica: Usuários de banda larga x Estudantes concluindo ensino primário (normalizado)')

plt.legend()
plt.xticks(rotation=90)
plt.show()
```

## Conclusões
Esperava encontrar uma relação entre o aumento de uso da internet com a quantidade de alunos concluindo o ensino primário, porém com a análise vimos que esses itens não estão relacionados. Tivemos um grande aumento de usuários de banda larga na Costa Rica enquanto o número de estudantes concluindo o ensino primário teve um aumento muito pequeno.
