# Examine graficamente as funções,

$$ y(t) = e^{-t} $$
$$ y(t) = e^{-t} cos(2 \pi t) $$
$$ y(t) = e^{-|t|} sin(t) $$

Qual delas apresenta evolução monotonicamente (sempre em um único sentido) decrescente? 




```python
# Seu código. 

# Dica: empregue esses valores de t para exibir as funções

plt.figure(figsize=(12,4)) # cria uma figura alterando o tamanho, figsize=(tamanho_horizontal,tamanho_vertical)
plt.suptitle('Funções', fontsize=14, weight='bold', y=1.05)

# gráfico 1
plt.subplot(1, 3, 1) # (linhas, colunas, gráfico ativo)
t = np.arange(-3,3,0.01)
plt.plot(t, np.exp(-t))
plt.title('$ y(t) = e^{-t} $')
plt.xlabel('t')
plt.ylabel('f(t)')

# gráfico 2
plt.subplot(1, 3, 2) # (linhas, colunas, gráfico ativo)
t = np.arange(-3,3,0.01)
plt.plot(t, np.exp(-t) * np.cos(4* np.pi * t))
plt.title('$ y(t) = e^{-t} cos(2 \pi t) $')
plt.xlabel('t')
plt.ylabel('f(t)')

# gráfico 3
plt.subplot(1, 3, 3) # (linhas, colunas, gráfico ativo)
t = np.arange(-5,5,0.01)
plt.plot(t, np.exp(-abs(t)) * np.sin(t))
plt.title('$ y(t) = e^{-|t|} sin(t) $')
plt.xlabel('t')
plt.ylabel('f(t)')

plt.tight_layout()
plt.show()
```

```python
t = np.arange(0,1,0.01)
plt.plot(t, np.exp(-t), label='$ y(t) = e^{-t} $')
plt.plot(t, np.exp(-t) * np.cos(4* np.pi * t),label='$ y(t) = e^{-t} cos(2 \pi t) $')
plt.plot(t, np.exp(-abs(t)) * np.sin(t), label='$ y(t) = e^{-|t|} sin(t) $')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)
#r = np.random.rand(1,10,1000)
r = np.random.rand(1,10000,1)

plt.figure(figsize=(15,8))

plt.scatter(r, np.e**-r, color="tab:cyan", s=10, marker='^', label='1')
plt.scatter(r, np.e**-r * np.cos(2 * np.pi * r), color="tab:purple", s=10, label='2')
plt.scatter(r, np.e**-np.abs(r) * np.sin(r), color="tab:green", s=10, label='3')
plt.legend()
```

# Examine a distribuição de idades dos passageiros de `Titanic`,

Graficamente você identifica que tipo de distribuição de idades? 

```python
titanic = sns.load_dataset('titanic')
display(titanic.head())

# Seu código

titanicHomens=titanic [ titanic.sex == 'male' ]
titanicMulheres=titanic [ titanic.sex == 'female' ]

plt.figure(figsize=(12,4)) # cria uma figura alterando o tamanho, figsize=(tamanho_horizontal,tamanho_vertical)

plt.subplot(1,2,1)
plt.hist(titanic.age, alpha=0.6, label='geral')
plt.hist(titanicHomens.age, alpha=0.6, label='homens')
plt.hist(titanicMulheres.age, alpha=0.6, label='mulheres')
plt.title('Distribuição de idades',fontsize=14,weight='bold')
plt.legend()


plt.subplot(1,2,2)
sns.kdeplot(titanic.age, alpha=0.6, label='geral')
sns.kdeplot(titanicHomens.age, alpha=0.6, label='homens')
sns.kdeplot(titanicMulheres.age, alpha=0.6, label='mulheres')
plt.title('Distribuição de idades - Seaborn',fontsize=14,weight='bold')
plt.legend()

plt.tight_layout()
plt.show()
```

# Tendência de Valores da Bolsa,

Analisando graficamente os índices da Bolsa da base abaixo, qual ou quais índices apresentam tendência (movimentos de alta e baixa) discrepantes com relação ao IBOV no período? 

```python
import pandas as pd
df = pd.read_csv('PLP_Python_3_bovespa2020_2021.csv')
display(df.head())

# Seu código
# Dica: Normalize os valores de cada índice, por exemplo fazendo
# df.valor = df.valor / max(df.valor)
df.IBOV=df.IBOV/max(df.IBOV)
df.VALE3=df.VALE3/max(df.VALE3)
df.PETR4=df.PETR4/max(df.PETR4)
df.DOLAR=df.DOLAR/max(df.DOLAR)

# Gráficos
plt.figure(figsize=(12,4))

plt.subplot(1,1,1)
plt.title('Movimentação Bolsa (normalizado)',fontsize=14,weight='bold')
plt.plot(df.data, df.IBOV, label='IBOV')
plt.plot(df.data, df.VALE3, label='VALE3')
plt.plot(df.data, df.PETR4, label='PETR4')
plt.plot(df.data, df.DOLAR, label='DOLAR')
plt.legend()

plt.tight_layout()
plt.show()
```
