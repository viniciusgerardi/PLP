##Primeiras impressões

Vamos inicializar com uma impressão básica da área de plotagem, utilizando duas variáveis para definir as áreas que serão condizentes à amplitude dos eixos **X** e **Y** do gráfico em 2D.

Repare que, o eixo definido como X é da variável *Price*, a escala seguirá esta amplitude. Da mesma forma, observe a escala (tamanho mínimo e máximo) do eixo Y.

```r
library(ggplot2)

ggplot(diamonds, aes(x=price, y=carat)) 
```

Ao inverter as variáveis, confira novamente os eixos X e Y para ver o resultado do tamanho da área de plotagem.

```r
ggplot(diamonds, aes(x=carat, y=price)) 
```

Neste momento a impressão foi apenas da área, sem nenhum elemento visual que vai nos mostrar alguma informação dos dados. Podemos então, adicionar um elemento de objetos geométrico para que represente o nosso dado no gráfico. A forma mais simples é utilizar o *geom_point()*, que permitirá visualizar um gráfico de dispersão (pontos) a partir da junção das duas variáveis no plano cartesiano. Veja o código abaixo a saída que o modelo apresenta.

```r
ggplot(diamonds, aes(x=carat, y=price)) + geom_point()
```

Repare que a saída é um gráfico de dispersão básico, onde cada ponto representa uma junção entre *Carat* (eixo X) e *Price* (eixo Y). No entando, a maioria dos pontos estão concentrados na parte esquerda e inferior do gráfico, isso acontece porque há uma maior concentração de diamantes que são mais leves e custam menos. Com o passar dos exercícios vamos explorar cada vez mais estas características.

Para deixar o gráfico auto explicativo para quem for consumir, é importante ajustar os nomes dos componentes. Para fazer isso, vamos adicionar o título ao gráfico e também rótulos para os eixos X e Y. Para questões pedagógicas, vamos utilizar as funções separadas para *ggtitle()*, *xlab()* e *ylab()*, que representam o ajuste do título e em seguida dos rótulos do eixo X e Y. Veja o código como fica.

```r
ggplot(diamonds, aes(x=carat, y=price)) +
  geom_point() +
  ggtitle("Análise de Diamantes", subtitle="Peso X Preço") +
  xlab("Peso do diamante") +
  ylab("Preço do Diamante")
```

É possível ajustar a camada estética de um elemento geométrico modificando os respectivos parâmetros. Vamos mudar a cor dos pontos e também o seu tamanho. Repare no código que está dentro do *geom_point()*. O parâmetro *col* define a cor e o parâmetro *size* define o tamanho.

```r
ggplot(diamonds, aes(x=carat, y=price)) +
  geom_point(col="darkcyan", size=1) +
  ggtitle("Análise de Diamantes", subtitle="Peso X Preço") +
  xlab("Peso do diamante") +
  ylab("Preço do Diamante")
```

Já que é possível ajustar os valores do elemento geométrico a partir de valores fixos, o que será que acontece se passarmos variáveis para estes parâmetros? Vamos ver como fica este elemento gráfico ao utilizarmos a variável *carat* no tamanho dos pontos impressos. Veja este bloco de código.

```r
ggplot(diamonds, aes(x=carat, y=price)) +
  geom_point(col="darkcyan", aes(size=carat)) +
  ggtitle("Análise de Diamantes", subtitle="Peso X Preço") +
  xlab("Peso do diamante") +
  ylab("Preço do Diamante")
```

Da mesma forma que foi possível definir o tamanho do objeto impresso no gráfico, será que conseguimos explorar também a impressão de cores a partir de uma variável?
Ao incluir a variável *Clarity* no parâmetro *col*, a aparência do gráfico fica como a impressão a seguir.
Repare que, tanto o tamanho dos pontos quanto a cor estão variando de acordo com a variável que foi utilizada em sua impressão.

```r
ggplot(diamonds, aes(x=carat, y=price)) +
  geom_point(aes(col=clarity, size=carat)) +
  ggtitle("Análise de Diamantes", subtitle="Peso X Preço") +
  xlab("Peso do diamante") +
  ylab("Preço do Diamante")
```

Para chegar ao fim desta exploração inicial, podemos alterar a aparência da área de plotagem inteira usando temas predefinidos. A página de ajuda mostra todos os temas integrados disponíveis. Assim como estão disponíveis lá no arquivo de *telas interativas* alguns exemplos de temas utilizados em um gráfico.

```r
help(theme_bw)
```

Para ajustar o tema, é necessário incluir à instrução do *ggplot* a configuração geral do tema (por exemplo *theme_bw()*). 

```r
ggplot(diamonds, aes(x=carat, y=price)) +
  geom_point(aes(col=clarity, size=carat)) +
  ggtitle("Análise de Diamantes", subtitle="Peso X Preço") +
  xlab("Peso do diamante") +
  ylab("Preço do Diamante") +
  theme_bw()
```

## Visualizações mais avançadas

Como vimos na seção de **Primeiras Impressões** as camadas em *ggplot2* são conhecidas como *geoms*. Assim que a configuração da base do seu gráfico estiver concluída, você pode adicionar novos **geoms** um em cima do outro.

Neste caso, veja que adicionamos duas camadas a este gráfico. Inicialmente, lá no primeiro gráfico que construímos, era só o geom_point() e agora, adicionamos uma nova camada, geom_smooth().

Em gráficos de dispersão pode ser difícil de visualizar tendências, principalmente porque são apenas pontos impressos. Muitas vezes, para ajudar neste sentido, desejamos adicionar uma linha de suavização para ver como são as tendências. Isso pode ser especialmente útil ao tentar entender as regressões (estimação de valores numéricos). O geom_smoth ajuda exatamente para isso, ele adiciona uma média condicional suavizada nos dados.



```r
ggplot(diamonds, aes(x=carat, y=price)) + geom_point() + geom_smooth()
```

Elaborando um pouco mais este gráfico, veja que o eixo X, o eixo Y e a cor foram definidos na configuração da própria *ggplot()*, essas duas camadas herdaram a estética.
Alternativamente, você pode especificar a estética dentro da camada *geom* como mostrado abaixo. E neste caso, ao herdar os elementos de estética, cada série da *Clarity* montará uma linha de tendência com a suavização.

```r
ggplot(diamonds, aes(x=carat, y=price)) +
  geom_point(aes(col=clarity), size=1) +
  ggtitle("Análise de Diamantes", subtitle="Peso X Preço") +
  xlab("Peso do diamante") +
  ylab("Preço do Diamante") +
  geom_smooth(aes(x=carat, y=price, col=clarity))
```

A visualização não ficou tão simples de enxergar com todos estes dados amontoados. Nestes casos, uma alternativa elegante de gerar a visualização é separar os dados em faces, onde cada face representará um sub-grupo do conjunto de dados. Esta segmentação é realizada com o elemento *facet_wrap*, que recebe o nome da variável com a qual se deseja fazer a segmentação dos dados. Repare no código abaixo, ao informar a variável *Clarity* no elemento de *facet_wrap*, os dados são segmentados em cada um dos valores existentes na variável *Clarity* e então o gráfico é feito utilizando cada um destes sub-grupos. Veja o resultado deste gráfico.


```r
ggplot(diamonds, aes(x=carat, y=price)) +
  geom_point(aes(col=clarity), size=1) +
  ggtitle("Análise de Diamantes", subtitle="Peso X Preço") +
  xlab("Peso do diamante") +
  ylab("Preço do Diamante") +
  geom_smooth(aes(x=carat, y=price, col=clarity)) +
  facet_wrap( ~ clarity) 
```

## Paleta de cores

Neste e-book você viu sempre as mesmas cores de gráficos porque elas são da paleta de cores padrão da biblioteca *ggplot2*. Se você se lembrar, em outros materiais aqui do componente curricular, já usamos cores diferentes. Nós podemos fazer isso também no *ggplot2*! No R, uma das bibliotecas que fornece paletas de cores alternativas é a RColorBrewer. Nós já usamos antes, e vamos utilizar agora de novo.

O caminho é simples, veja:

- Carregar a biblioteca *RColorBrewer*;
- Imprima a lista de paletas disponíveis com a função *display.brewer.all()*;
- Ao decidir sua paleta de cores, aplique ao gráfico;

Para iniciar, vamos ver como é o carregamento da biblioteca e então, ver sua paleta de cores. Acompanhe o código abaixo:


```r
library(RColorBrewer)
display.brewer.all()
```

Veja um exemplo de gráfico que explora essa base de dados. Espero que você já consiga interpretar os elementos utilizado neste gráfico:

```r
library(ggplot2)
library(tidyverse)

graficoDiamantes <-  ggplot(diamonds, aes(x = color, fill = clarity)) + 
    geom_bar(position = 'fill') +
    xlab('Cores do diamante \n de D (melhor) até J (pior)') +
    ylab('Proporções') +
    labs(fill = "Claridade do diamante \n de I1 (pior) até IF (melhor)") +
    facet_grid(. ~ cut) + 
    ggtitle("Detalhes dos diamantes!")

graficoDiamantes
```

Como a variável que estamos utilizando para preencher o gráfico é a *clarity*, e ela é categórica, utilizaremos a função *scale_fill_brewer* passando o nome da paleta que queremos ver nos dados.

um detalhe importante é que para utilizar a paleta em variáveis categóricas, a função é: *scale_fill_brewer(palette = '<<nome da paleta>>')* enquanto que, para utilizar em variáveis numéricas, deve ser utilizada a função *scale_fill_distiller(palette = '<<nome da paleta>>')*.

Agora veja o mesmo gráfico utilizando cores da paleta *Spectral*:

```r
graficoDiamantes + scale_fill_brewer(palette = 'Spectral')
```
